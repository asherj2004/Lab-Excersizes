```python
import requests
import tkinter as tk
from tkinter import ttk

API_URL = 'https://api.quotable.io'

def get_authors():
    response = requests.get(f'{API_URL}/authors')
    data = response.json()
    return data['results']

def get_quote_by_author(author):
    response = requests.get(f'{API_URL}/random?author={author}')
    data = response.json()
    return data['content']

def display_quote():
    author = author_var.get()
    quote = get_quote_by_author(author)
    quote_label.config(text=quote)

def update_authors():
    authors = get_authors()
    author_var.set('')
    author_menu['menu'].delete(0, 'end')
    for author in authors:
        author_menu['menu'].add_command(label=author['name'], command=tk._setit(author_var, author['name']))

root = tk.Tk()
root.title('Quotable')

author_label = ttk.Label(root, text='Select an author:')
author_label.pack(pady=10)
author_var = tk.StringVar()
author_menu = ttk.OptionMenu(root, author_var, '', '')
author_menu.pack()

quote_button = ttk.Button(root, text='Get Quote', command=display_quote)
quote_button.pack(pady=10)

quote_label = ttk.Label(root, wraplength=400)
quote_label.pack(pady=10)

update_button = ttk.Button(root, text='Update Authors', command=update_authors)
update_button.pack(pady=10)

update_authors()

root.mainloop()
```

```python
import tkinter as tk
import requests

API_ENDPOINT = 'https://api.weather.gov/gridpoints/SEW/141,64/forecast'

def fetch_forecast():
    response = requests.get(API_ENDPOINT)
    data = response.json()
    return data['properties']['periods']

root = tk.Tk()
root.title('7-Day Forecast')

location_label = tk.Label(root, text='Seattle, WA')
location_label.pack()

current_temp_label = tk.Label(root, font=('Arial', 48), text='--°F')
current_temp_label.pack()

current_conditions_label = tk.Label(root, text='--')
current_conditions_label.pack()

separator = tk.Frame(height=2, bd=1, relief=tk.SUNKEN)
separator.pack(fill=tk.X, padx=5, pady=5)

forecast_frame = tk.Frame(root)
forecast_frame.pack()

forecast = fetch_forecast()

for day in forecast:
    # this is a frame, which is a widget we did not use in class
    day_frame = tk.Frame(forecast_frame)
    day_frame.pack(side=tk.LEFT, padx=10)

    day_name_label = tk.Label(day_frame, font=('Arial', 16), text=day['name'])
    day_name_label.pack()

    day_temp_label = tk.Label(day_frame, font=('Arial', 12), text=f"{day['temperature']}°F")
    day_temp_label.pack()

    day_conditions_label = tk.Label(day_frame, text=day['shortForecast'])
    day_conditions_label.pack()

root.mainloop()
```
